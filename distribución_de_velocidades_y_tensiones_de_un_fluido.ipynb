Grafica de tensiones de corte junto con el perfil de velocidades de agua entre dos cilindro concentricos usando Python

```python
Este gráfico muestra:
- En azul: cómo cambia la velocidad axial Vr𝑧(𝑟) desde el cilindro interior hasta el exterior.
- En rojo: cómo varía la tensión de corte 𝜏 𝑟𝑧 (𝑟), que disminuye al aumentar el radio.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
mu = 1e-3  # viscosidad dinámica (Pa·s)
R1 = 0.01  # radio interno (m)
R2 = 0.015  # radio externo (m)
C1 = -0.04934  # constante calculada
C2 = -0.207    # constante calculada

# Vector de radios
r = np.linspace(R1, R2, 300)

# Perfil de velocidad axial
vz = C1 * np.log(r) + C2

# Tensión de corte: tau_rz = -mu * (dvz/dr) = -mu * (C1 / r)
tau_rz = -mu * (C1 / r)

# Graficar
fig, ax1 = plt.subplots(figsize=(8, 5))

# Eje para velocidad
ax1.plot(r * 1000, vz, 'b-', label='Velocidad $v_z(r)$')
ax1.set_xlabel('Radio [mm]')
ax1.set_ylabel('Velocidad [m/s]', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Segundo eje para tensión de corte
ax2 = ax1.twinx()
ax2.plot(r * 1000, tau_rz, 'r--', label='Tensión de corte $\\tau_{rz}(r)$')
ax2.set_ylabel('Tensión de corte [Pa]', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Líneas en los radios R1 y R2
ax1.axvline(R1*1000, color='gray', linestyle=':', lw=1)
ax1.axvline(R2*1000, color='gray', linestyle=':', lw=1)

# Título y leyendas
fig.suptitle('Distribución de velocidad y tensión de corte entre cilindros')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.grid(True)
plt.tight_layout()
plt.show()
```
