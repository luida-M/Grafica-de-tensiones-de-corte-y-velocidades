Grafica de tensiones de corte junto con el perfil de velocidades de agua entre dos cilindro concentricos usando Python

```python
Este gr치fico muestra:
- En azul: c칩mo cambia la velocidad axial Vr洧녾(洧) desde el cilindro interior hasta el exterior.
- En rojo: c칩mo var칤a la tensi칩n de corte 洧랦 洧洧녾 (洧), que disminuye al aumentar el radio.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
mu = 1e-3  # viscosidad din치mica (Pa췅s)
R1 = 0.01  # radio interno (m)
R2 = 0.015  # radio externo (m)
C1 = -0.04934  # constante calculada
C2 = -0.207    # constante calculada

# Vector de radios
r = np.linspace(R1, R2, 300)

# Perfil de velocidad axial
vz = C1 * np.log(r) + C2

# Tensi칩n de corte: tau_rz = -mu * (dvz/dr) = -mu * (C1 / r)
tau_rz = -mu * (C1 / r)

# Graficar
fig, ax1 = plt.subplots(figsize=(8, 5))

# Eje para velocidad
ax1.plot(r * 1000, vz, 'b-', label='Velocidad $v_z(r)$')
ax1.set_xlabel('Radio [mm]')
ax1.set_ylabel('Velocidad [m/s]', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Segundo eje para tensi칩n de corte
ax2 = ax1.twinx()
ax2.plot(r * 1000, tau_rz, 'r--', label='Tensi칩n de corte $\\tau_{rz}(r)$')
ax2.set_ylabel('Tensi칩n de corte [Pa]', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# L칤neas en los radios R1 y R2
ax1.axvline(R1*1000, color='gray', linestyle=':', lw=1)
ax1.axvline(R2*1000, color='gray', linestyle=':', lw=1)

# T칤tulo y leyendas
fig.suptitle('Distribuci칩n de velocidad y tensi칩n de corte entre cilindros')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.grid(True)
plt.tight_layout()
plt.show()
```
